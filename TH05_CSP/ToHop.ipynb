- A gồm N kí tự và K số nguyên
- {A,B}={B,A}
Ý tưởng
+ Lưu trữ tất cả các kí hiệu vào mảng
+ duyệt từ 0 đến K-1 để lấy ra kí tự để chọn
+ đảm bảo không còn phần tử nào phía trước
+ điều kiện dừng khi chọn đủ K phần tử  
+ Ở mỗi bắt lặp bắt đầu từ Start đến N-(N-index) nếu vượt quá không còn phần tử nào để 

```python
import os


def backtrack(start, n, k, a, current, result):
    if len(current) == k:
        result.append(" ".join(current))
        return
    for i in range(start, n):
        current.append(a[i])
        # Truyen tat ca cac bien vao ham de quy
        backtrack(i + 1, n, k, a, current, result)
        current.pop()  # Quay lui


def solve():
    file_name = "Tohop.txt"
    if not os.path.exists(file_name):
        print(f"Không tìm thấy file {file_name}")
        return

    with open(file_name, "r") as f:
        all_data = f.read().split()

    if not all_data:
        return

    n = int(all_data[0])
    k = int(all_data[1])

    # bat dau tu index 2 trong all_data
    a = all_data[2 : 2 + n]

    result = []
    current = []
    backtrack(0, n, k, a, current, result)

    print(len(result))
    for res in result:
        print(res)


solve()
```
