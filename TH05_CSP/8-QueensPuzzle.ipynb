- Đặt các ô lần lượt theo từng cột từ cột 1 đến cột 8
- Điều kiện để đặt
+ không có quân hậu nào trên cùng hàng
+ không có quân hậu khác trên cả hai đường chéo
- Bước thực hiện:
+ Bắt đầu với cột đầu tiên
+ thử đặt hậu vào từ hàng từ 1 đến 8 trong cột đó
+ nêu vị trí an toàn thì đánh dấu và đệ quy để tìm ô tiếp theo
+ nếu đặt quan hậu ở cột thứ 8, ta tìm được lời giải
+ nếu không tìm được hàng an toàn ở ô hiện tại , quay lùi bỏ quân hậu ở cột trước và thử hàng tiếp 


```python
import os

def backtracking_queen():
    n = 8
    diemdung = [-1] * n
    cot = [False] * n
    hangcheochinh = [False] * (2 * n - 1)
    hangcheophu = [False] * (2 * n - 1)
    solutions = [] # Luu ket qua

    def backtracking(row):
        if row == n:
            board_res = []
            for r in range(n):
                line = ["."] * n
                line[diemdung[r]] = "Q"
                board_res.append(" ".join(line))
            solutions.append(board_res)
            return

        for col in range(n):
            if (not cot[col] and 
                not hangcheochinh[row + col] and 
                not hangcheophu[row - col + n - 1]):
                
                diemdung[row] = col
                cot[col] = hangcheochinh[row + col] = hangcheophu[row - col + n - 1] = True

                backtracking(row + 1)

                # Quay lui
                cot[col] = hangcheochinh[row + col] = hangcheophu[row - col + n - 1] = False

    backtracking(0)

    if solutions:
        print(f"Ket qua")
        for row in solutions[0]:
            print(row)
    else:
        print("Khong tim thay ket qua")

backtracking_queen()
```
