Ý tưởng Sudoku
- tìm ô còn trống trong bảng 9X9
- thử lần lượt từ 1-9 vào ô đó
- kiểm tra tính hợp lệ:
+ không trùng với ô cùng hàng
+ không trùng với cột
+ không trùng với các ô 3X3 tương
- Nếu só hợp lệ , tiếp tục đệ quy để điền số tiếp theo
+ nếu điềm hết tất các các ô trống trong bảng mà không vi phạm thì tìm được lời giải
+ nếu từ 1-9 không có số nào hợp lệ cho cho ô đó hoặc điền tiếp bị lỗi , ta quay lui bằng cách xóa số vừa điền (trả về rỗng) và thử số tiếp  theo ở bước trước
+ Nếu đã thử mọi khả năng mà không có lời giải , thì thông báo 


```python
import sys


def solve():
    file_path = "Sudoku.txt"
    try:
        with open(file_path, "r") as f:
            lines = [line.strip() for line in f.readlines() if line.strip()]
    except FileNotFoundError:
        print("File not found")
        return

    # Tao board
    board = []
    for line in lines:
        board.append(list(line))

    if len(board) != 9:
        print("Khong hop le")
        return

    # kiem tra
    def valid(r, c, val):
        for i in range(9):
            if board[r][i] == val or board[i][c] == val:
                return False

        start_r, start_c = 3 * (r // 3), 3 * (c // 3)
        for i in range(start_r, start_r + 3):
            for j in range(start_c, start_c + 3):
                if board[i][j] == val:
                    return False
        return True

    def backtrack():
        for r in range(9):
            for c in range(9):
                if board[r][c] == ".":
                    for val in "123456789":
                        if valid(r, c, val):
                            board[r][c] = val
                            if backtrack():
                                return True
                            board[r][c] = "."
                    return False
        return True
    
    if backtrack():
        for row in board:
            print("".join(row))
    else:
        print("IMPOSSIBLE")


# chay
solve()
```
