Thuật toán tô màu:
- Sử dụng chiến lược thử sai
- thử tô toàn bộ bằng 1 màu k=1
- nếu không thể tô được nữa thì tăng màu tô lên K=2
- lặp lại quá trình cho đến khi tìm được k nhỏ với hàm backtracking trả về

Mô tả backtracking:
- Giả sử thử tô màu với K màu tối đa
+ Điều kiện dừng , nếu đã tô cho đỉnh cuối cùng (Node =N+1) nghĩa là ta đã tìm thấy 1 cách tô hợp lệ,xuất ra kết quả và dừng chương trình
+ Lựa chọn màu: với đỉnh hiện tại , lần lượt tô các màu từ 1 đến k
+ kiểm tra hợp lê : màu C được coi là an toàn cho đỉnh U nếu tất cả cá đỉnh lân cận chưa được hoặc được tô có màu khác C
+ Đệ quy(tiến tới): Nếu màu C an toàn , gán C cho đỉnh U và gọi đệ quy cho các điểm kế tiếp (U+1)
+ Quay lui: Nếu không còn màu nào từ 1 đến K phù hợp để tô cho đỉnh tiếp theo, ta xóa màu của đỉnh U( gán bằng 0) , rồi quay lại bước trước để tìm màu khác cho đỉnh cũ

```python
import os


def is_safe(node, color_to_try, adj, color):
    # Kiem tra xem co hang xom nao cua 'node' da to mau trycolor
    for neighbor in adj[node]:
        if color[neighbor] == color_to_try:
            return False
    return True


def backtrack(node, k, n, adj, color):
    # Neu da to du n thanh pho
    if node > n:
        return True

    # To mau  tu 1 den k
    for c in range(1, k + 1):
        if is_safe(node, c, adj, color):
            color[node] = c
            if backtrack(node + 1, k, n, adj, color):
                return True
            color[node] = 0
    return False


def solve():
    file_name = "Tomau.txt"

    if not os.path.exists(file_name):
        print(f"Không tìm thấy file {file_name}")
        return

    with open(file_name, "r") as f:
        inputdata = f.read().split()

    if not inputdata:
        return

    # n va m la hai thanh pho ke nhau
    n = int(inputdata[0])
    m = int(inputdata[1])

    # xay dung danh sach ke
    adj = [[] for _ in range(n + 1)]
    idx = 2
    for _ in range(m):
        if idx + 1 < len(inputdata):
            u = int(inputdata[idx])
            v = int(inputdata[idx + 1])
            adj[u].append(v)
            adj[v].append(u)
            idx += 2

    final_color = [0] * (n + 1)
    min_k = 0

    #thu tang k tu 1 den n de to mau it nhat
    for k in range(1, n + 1):
        color = [0] * (n + 1)
        if backtrack(1, k, n, adj, color):
            min_k = k
            final_color = color
            break

    print(min_k)
    for color_idx in range(1, min_k + 1):
        group = []
        for i in range(1, n + 1):
            if final_color[i] == color_idx:
                group.append(str(i))

        if group:
            print(f"Màu {color_idx}: {' '.join(group)}")


solve()
```
