Ý tưởng :
- Mã đi qua tất cả các ô đúng 1 lần
- xét trạng thái (x,y)
- điều kiện dừng khi đủ bước n*n
- Mã có tối đa 8 cách di chuyển theo hình chữ L
+ tại mỗi bước thử một trong ô có thể đi tiếp 
+ Điều kiện hợp lệ : phải nằm trong bàn cờ vua và chưa được đi qua
+ nếu hợp lệ đánh đấu ô đã đi và đệ quy để  tìm ô tiếp theo
+ nếu đi đủ n x n bước thì kết thúc
+ nếu tại một vị trí mà 8 hướng đều không co lời giải, xóa đi dấu viết của ô hiện tại và quay về vị trí trước đó để thử hướng khác

```python
def solve_knight():
    with open("Knight.txt", "r") as f:
        data = f.read().split()
    n = int(data[0])
    start_x, start_y = int(data[1]), int(data[2])

    # Khoi tao gia tri 0 cho o chua di
    board = [[-1 for _ in range(n)] for _ in range(n)]

    board[start_x][start_y] = 0

    # 8 Huong di chuyen
    dx = [2, 1, -1, -2, -2, -1, 1, 2]
    dy = [1, 2, 2, 1, -1, -2, -2, -1]

    if backtrack_knight(start_x, start_y, 1, n, board, dx, dy):
        for row in board:
            print(" ".join(f"{item:2}" for item in row))
    else:
        print("Không có lời giải")


def backtrack_knight(x, y, move_i, n, board, dx, dy):
    if move_i == n * n:
        return True

    for i in range(8):
        next_x = x + dx[i]
        next_y = y + dy[i]
        if 0 <= next_x < n and 0 <= next_y < n and board[next_x][next_y] == -1:
            board[next_x][next_y] = move_i
            if backtrack_knight(next_x, next_y, move_i + 1, n, board, dx, dy):
                return True
            board[next_x][next_y] = -1  # Quay lui
    return False


solve_knight()
```
