```python
def backtracking_queen():
    n = 8
    diemdung = [-1] * n
    cot = [False] * n
    hangcheochinh = [False] * (2 * n - 1)
    hangcheophu = [False] * (2 * n - 1)

    def backtracking(row):
        if row == n:
            print(diemdung)
            return

        for col in range(n):
            if (
                not cot[col]
                and not hangcheochinh[row + col]
                and not hangcheophu[row - col + n - 1]
            ):
                diemdung[row] = col
                cot[col] = True
                hangcheochinh[row + col] = True
                hangcheophu[row - col + n - 1] = True

                backtracking(row + 1)

                cot[col] = False
                hangcheochinh[row + col] = False
                hangcheophu[row - col + n - 1] = False

    backtracking(0)


backtracking_queen()
```
