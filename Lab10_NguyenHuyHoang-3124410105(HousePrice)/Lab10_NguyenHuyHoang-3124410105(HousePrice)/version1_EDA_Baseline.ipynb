```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Nạp dữ liệu thô
df_train = pd.read_csv('data/train.csv')
print("Dữ liệu train có:", df_train.shape)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df_train['SalePrice'], kde=True, color='blue', bins=50)
plt.title('Biểu đồ phân phối Giá nhà (SalePrice Distribution)')
plt.savefig('../outputs/01_price_distribution.png')
plt.show()
```

```python
# Tính độ tương quan
corr = df_train.select_dtypes(include=['number']).corr()
top_10_corr = corr.nlargest(10, 'SalePrice')['SalePrice'].index

plt.figure(figsize=(12, 10))
sns.heatmap(df_train[top_10_corr].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Top 10 yếu tố tương quan cao nhất với Giá nhà')
plt.savefig('outputs/02_heatmap.png')
plt.show()
```
