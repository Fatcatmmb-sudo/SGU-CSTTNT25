```python
import networkx as nx
import matplotlib.pyplot as plt

# Khởi tạo đồ thị
G = nx.Graph()
# A -> {B,C,F}, B -> {A,C,E,F}, C -> {A,B,D,F}, D -> {C,F}, E -> {B,D,F}, F -> {A,C,E}
edges = [
    ("A", "B"),
    ("A", "C"),
    ("A", "F"),
    ("B", "C"),
    ("B", "E"),
    ("B", "F"),
    ("C", "D"),
    ("C", "F"),
    ("D", "F"),
    ("E", "D"),
    ("E", "F"),
]
G.add_edges_from(edges)



```

# Tính bậc của từng đỉnh

```python
print("--- Bậc của từng đỉnh ---")
degrees = dict(G.degree())
max_degree = 0
nodes_max_degree = []

for node, deg in degrees.items():
    print(f"Đỉnh {node}: bậc {deg}")
    if deg > max_degree:
        max_degree = deg
```

# Tìm các đỉnh có bậc cao nhất

```python
nodes_max_degree = [node for node, deg in degrees.items() if deg == max_degree]
print(f"\nBậc cao nhất là {max_degree} tại các đỉnh: {', '.join(nodes_max_degree)}")
```

# Tìm đường đi từ A đến D (Ngắn nhất)

```python
shortest_path = nx.shortest_path(G, source="A", target="D")
print(f"\nĐường đi ngắn nhất từ A đến D: {' -> '.join(shortest_path)}")
```

# Vẽ đồ thị

```python

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)  # Định vị các đỉnh

nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="lightblue",
    node_size=2000,
    edge_color="gray",
    font_size=15,
    font_weight="bold",
)

plt.title("Biểu diễn đồ thị từ hình ảnh")
plt.show()
```
