Các Hàm xử lý

```python
def XuLyKhongTrung(a):
    s = set(a)
    return sorted(s)
```

```python
def DemSoLanXuatHien(a):
    dem = {}
    for ky_tu in a:
        if ky_tu in dem:
            dem[ky_tu] += 1
        else:
            dem[ky_tu] = 1
    return dem
```

```python
#test
input_str = input("Nhập vào một dãy ký tự (cách nhau bởi khoảng trắng): ")
a = input_str.split()
print("Dãy ký tự vừa nhập:", a)

danh_sach_loc = XuLyKhongTrung(a)
print("Ký tự không trùng (tăng dần):", danh_sach_loc)

tu_dien_dem = DemSoLanXuatHien(a)
print("Tần suất xuất hiện:", tu_dien_dem)
```
