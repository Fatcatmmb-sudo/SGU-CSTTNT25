Nhập một số nguyên n. Hãy tính tổng các số chẵn từ 1 đến  
Moi ban nhap so nguyen n: 10  
Tong cac so chan tu 1 den 100 la 2550.

# Nhập dữ liệu

```python

n = int(input("Moi ban nhap so nguyen n: "))
```

# Bài làm


```python
s = 0
for i in range(2, n + 1, 2):
    s += i
print(f"tong so chan tu 1 den {n} la {s}")
```
