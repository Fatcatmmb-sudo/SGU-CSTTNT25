```python
import math

def GiaiPhuongTrinhBac2(a, b, c):
    flag = None
    x = ()
    # Trường hợp a = 0: Trở thành phương trình bậc 1 bx + c = 0
    if a == 0:
        if b == 0:
            if c == 0:
                flag = -1  # Vô số nghiệm
                x = ()
            else:
                flag = 0  # Vô nghiệm
                x = ()
        else:
            flag = 1  
            x = (-c / b,)

    # Trường hợp a != 0: Phương trình bậc 2
    else:
        delta = b**2 - 4 * a * c
        if delta < 0:
            flag = 0
            x = ()
        elif delta == 0:
            flag = 1
            x = (-b / (2 * a),)
        else:
            flag = 2
            x1 = (-b + math.sqrt(delta)) / (2 * a)
            x2 = (-b - math.sqrt(delta)) / (2 * a)
            x = (x1, x2)

    return flag, x
```
