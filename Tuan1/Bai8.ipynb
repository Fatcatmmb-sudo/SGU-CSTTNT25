Nhap Mang

```python
def NhapMang():
    s = input("moi nhap mang cach nhau boi khoang trang: ")
    a = [float(x) for x in s.split()]
    return a
```

Ham Sinh Ngau nhien

```python
import random
def TaoMangNgauNhien(n,vmin=-10,vmax=10):
    a=[]
    for i in range(n):
        a.append(random.randint(vmin,vmax))
        return a
    
```

Ham dem tong so chan, le

```python
def DemTongChanLe(a):
    tong=sum(a)
    tongchan=0
    tongle=0
    for x in a:
        if x%2==0:
            tongchan+=x
        else:
            tongle+=x
    return tong,tongchan,tongle
```

Xuat Chan le

```python
def DayChanLe(a):
    daychan=[]
    dayle=[]
    for x in a:
        if x%2==0:
            daychan.append(x)
        else:
            dayle.append(x)
    return daychan,dayle
```

```python

a = NhapMang()
print("Mang ban vua nhap la: ", a)

tong_tat_ca, so_chan, so_le = DemTongChanLe(a)
ds_chan, ds_le = DayChanLe(a)

print("Tong tat ca cac so trong mang la: ", tong_tat_ca)
print("So luong so chan la: ", so_chan)
print("So luong so le la: ", so_le)
print("Cac so chan trong mang la:", ds_chan)
print("Cac so le trong mang la:", ds_le)
```
