Bài 1. Tính giá trị hàm số
Hãy viết chương trình nhập vào số thực x và tính giá trị biểu thức:

Bài Làm

```python
import math
```

Nhập dữ liệu và ép kiểu
dùng input() để đưa dữ liệu vài và éo thành kiểu số nguyên Int()

```python
x= int(input("nhập giá trị của biến x:"))
```

thành phần 1

```python
p1=x
```

thành phần 2: x^5/5!

```python
p2=(x**5)/math.factorial(5)
```

thành phần 3: sqrt(|x|)/x^(3/2)

```python
p3=math.sqrt(math.fabs(x))/(x**1.5)
```

Tính tổng và xuất ra màn hình

```python
fx=p1+p2+p3
print(f"Gia tri cua f({x}) la {fx:.2f}")
```
